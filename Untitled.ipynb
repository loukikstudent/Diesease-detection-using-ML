```python
def probTarget(list,a):
    b=list.count(a)
    n=len(list)
    prob=(float(b)/n)
    return prob
```

```python
def allProbTarget(list):
    targetProb=[]
    for i in range (len(list)):
        targetProb.append(probTarget(list,list[i]))
    return targetProb
```
